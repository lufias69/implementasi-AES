```python
from os import listdir
from os.path import isfile, join
def dfile (path):
    mypath = path
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    return onlyfiles
```

```python
import pandas as pd
import numpy as np
```

# Load Kunci Jawaban

```python
df = pd.read_excel("data\kunci jawaban dosen\kunci_jawaban_dosen.xlsx")
kode = df.kode.tolist()
skor = df.skor.tolist()
kunci_jawaban = df.jawaban_jawaban.tolist()

dict_kunci_jawaban = dict(zip(kode, kunci_jawaban))
dict_skor = dict(zip(kode, skor))
dict_skor
```

# Load Jawaban 

```python
jawaban_file = dfile ("data/jawaban")
jawaban_file

dict_jawaban_file = dict(zip(kode, jawaban_file))
dict_jawaban_file
```

# Load Library AES

```python
from aes_class import AES
hasil_dict = dict()
```

# Parameter

```python
tf=True
# qe=False
ngram=True
n=1
stemming=True
binary = False

# tf=tf, qe=qe, ngram=ngram, n=n, stemming=stemming, binary = binary
```

```python

def percobaan(kode):
    #tanpa QE
    kode = kode
    dickt = dict()
    model=  AES.AES(tf=tf, qe=False, ngram=ngram, n=n, stemming=stemming, binary = binary)#otak atik disini
    model.train([dict_kunci_jawaban[kode]], [dict_skor[kode]])

    df = pd.read_excel("data/jawaban/"+dict_jawaban_file[kode])
    jawaban = df.jawaban.tolist()


    sim_list = list()
    for j in jawaban:
        model.predict(j)
        similarity = model.hasil_cosine_similarity
        sim_list.append(similarity[0])
        print(".",end=".")
    hasil_dict.update({kode:sim_list})
    dickt.update({kode:sim_list})

    #Menggunakan QE
    model=  AES.AES(tf=tf, qe=True, ngram=ngram, n=n, stemming=stemming, binary = binary)#otak atik disini
    model.train([dict_kunci_jawaban[kode]], [dict_skor[kode]])

    df = pd.read_excel("data/jawaban/"+dict_jawaban_file[kode])
    jawaban = df.jawaban.tolist()


    sim_list = list()
    for j in jawaban:
        model.predict(j)
        similarity = model.hasil_cosine_similarity
        sim_list.append(similarity[0])
        print(".",end=".")
    hasil_dict.update({kode+"_QE":sim_list})
    dickt.update({kode+"_QE":sim_list})
    return pd.DataFrame.from_dict(dickt)
```
